```python
# 自動車の属性から価格を予測するモデルを作ろう
# 重回帰分析　
# 目的変数:　price
# 説明変数: 今回はhorepower, width, heightの3つだけ使う

import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

%matplotlib inline
%precision 3
```

```python
autos = pd.read_csv('datasets/imports-85.data', header=None)
```

```python
autos.columns = [
    'symboling',
    'normalized-losses',
    'make',
    'fuel-type',
    'aspiration',
    'num-of-doors',
    'body-style',
    'drive-wheels',
    'engine-location',
    'wheel-base',
    'length',
    'width',
    'height',
    'curb-weight',
    'engine-type',
    'num-of-cylinders',
    'engine-size',
    'fuel-system',
    'bore',
    'stroke',
    'compression-ratio',
    'horsepower',
    'peak-rpm',
    'city-mpg',
    'highway-mpg',
    'price'
]
```

```python
auto_tmp = autos[['price','horsepower','width','height']]
auto_tmp.isin(['?']).sum()
```

```python
auto = auto_tmp.replace('?', np.nan).dropna()
auto.isin(['?']).sum()
```

```python
auto.dtypes
```

```python
auto.head()
```

```python
auto = auto.assign(price=pd.to_numeric(auto.price))
```

```python
auto = auto.assign(horsepower=pd.to_numeric(auto.horsepower))
```

```python
auto.dtypes
```

```python
auto.corr()
```

```python
X = auto.drop('price', axis=1)
y = auto['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5,random_state=0)

model = LinearRegression()
model.fit(X_train, y_train) # 学習させる
print("決定係数 train:{:.3f}".format(model.score(X_train, y_train)))
print('決定係数 test:{:.3f}'.format(model.score(X_test, y_test)))

print('回帰係数\n{}'.format(pd.Series(model.coef_, index=X.columns)))
print('切片\n{}'.format(model.intercept_))
```

```python
"""
Step1. 各種モデル構築のためのクラスをインスタンス化する
Step2. データを説明変数と目的変数に分ける
Step3. 訓練データとテストデータに分ける。
Step4. 訓練データによるあてはめ（fit）→学習させる
Step5. モデルの汎化性能をテストデータで確かめる

説明変数の選択方法
    変数増加法（前進的選択法）
    変数減少法（後退的選択法）
    ステップワイズ法

選択する基準
    RMSE（RootMeanSquaredError）
    赤池情報量基準（AIC）
    ベイズ情報用基準（BIC）

別途、他の教材で勉強する
"""
```

```python
auto = autos[['price','width','engine-size']]
auto = auto.replace('?', np.nan).dropna()
auto.isin(['?']).sum()
auto = auto.assign(price=pd.to_numeric(auto.price))
X = auto.drop('price', axis=1)
y = auto['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5,random_state=0)

model = LinearRegression()
model.fit(X_train, y_train) # 学習させる
print("決定係数 train:{:.3f}".format(model.score(X_train, y_train)))
print('決定係数 test:{:.3f}'.format(model.score(X_test, y_test)))

print('回帰係数\n{}'.format(pd.Series(model.coef_, index=X.columns)))
print('切片\n{}'.format(model.intercept_))
```

```python
def calc_entropy(p):
    return -(p * np.log2(p) + (1 - p) * np.log2(1- p))


p = np.arange(0.01, 0.999, 0.01)
plt.plot(p, calc_entropy(p))
plt.grid(True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X = mushroom_dummy.drop('classes_flg', axis=1)
y = mushroom_dummy['classes_flg']

X_train,X_test,y_train,y_test = train_test_split(X, y, random_state=0)

model = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=0)
model.fit(X_train,y_train)

print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

cancer_data = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(cancer_data.data, cancer_data.target, stratify= cancer_data.target, random_state=0)

training_acc = []
test_acc = []

for n_neighbors in range(1,21):
    model = KNeighborsClassifier(n_neighbors = n_neighbors)
    model.fit(X_train, y_train)
    training_acc.append(model.score(X_train, y_train))
    test_acc.append(model.score(X_test, y_test))

plt.plot(range(1,21), training_acc, label='Training')
plt.plot(range(1,21), test_acc, label='Test')
plt.ylabel('Acc')
plt.xlabel('n_neighbors')
plt.legend()
```
